# Data Cleaning Analysis

**Purpose:** Identify and assess data quality issues such as missing values, duplicates, and incorrect values.

**Tasks:**
1. Check for missing data (null values) and analyze the missingness pattern.
2. Check for duplicates (rows or columns).
3. Check for outliers using boxplots or statistical methods (e.g., IQR, Z-score).
4. Examine data inconsistencies, such as text formatting issues, incorrect data types, etc.
